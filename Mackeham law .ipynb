# PLOTTING THE SURVIVAL FUNCTION CURVES USING THE MACKEHAM LAW # 

```python
import numpy as np
import matplotlib.pyplot as plt

def survival_function(t, A, B, c):
    return np.exp(-A * t - B * (np.exp(c * t) - 1))

def probability_density_fxn(t, A, B, c):
    return A * np.exp(-A * t - B * (np.exp(c * t) - 1)) + B * c * np.exp(c * t) * np.exp(-A * t - B * (np.exp(c * t) - 1))

A = 0.0001
B = 0.00035
c = 1.075

# Age range from 0 to 130 years
age_range = np.linspace(0, 130, 1000)

# Calculate the survival function values
Sx_values = survival_function(age_range, A, B, c)

# Calculate the probability density function values
mx_values = probability_density_fxn(age_range, A, B, c)

# Plot the survival function Sx(t)
plt.figure(figsize=(10, 6))
plt.plot(age_range, Sx_values, label='Survival Function Sx(t)')
plt.xlabel('Age (years)')
plt.ylabel('Sx(t)')
plt.title('Survival Function of Makeham Law of Mortality')
plt.legend()
plt.grid(True)
plt.show()

# Plot the probability density function fx(t)
plt.figure(figsize=(10, 6))
plt.plot(age_range, mx_values, label='Probability Density Function fx(t)')
plt.xlabel('Age (years)')
plt.ylabel('fx(t)')
plt.title('Probability Density Function of Makeham Law of Mortality')
plt.legend()
plt.grid(True)
plt.show()

```
