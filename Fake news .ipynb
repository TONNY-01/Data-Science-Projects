```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Admin\Desktop\data.csv")
df
```

```python
df = df.dropna()
df
```

```python
y = df.Label
X = df.drop('Label', axis = 1)
X
```

```python
import tensorflow as tf
```

```python
voc_size = 30000
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
Messages = X.copy()
messages = Messages.reset_index()
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    news = re.sub('[^a-zA-Z]', ' ', messages['Body'][i])
    news = news.lower()
    news = news.split()
    
    news = [ps.stem(word) for word in news if not word in stopwords.words('english')]
    news = ' '.join(news)
    corpus.append(news)
```

```python
corpus
```

```python
onehot_rep=[one_hot(words, voc_size)for words in corpus]
onehot_rep
```

```python
sent_length=1000
embedded = pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded)
```

```python
embedding_vector_features = 100
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features, input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded)
y_final = np.array(y)
X_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)
```

```python
model.fit(X_train, y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
pred = model.predict(X_test)
pred = pred.round()
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print('accuracy: ', acc*100, '%')
#from sklearn.metrics import confusion_matrix
#confusion_matrix(y_test, pred)
```
